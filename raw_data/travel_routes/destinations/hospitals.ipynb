```python
import pandas as pd

pd.set_option('display.max_columns', None)

pd.set_option('display.max_rows', 100)

df = pd.read_csv('./raw_data/Hospitals.csv')

to_drop = [
    'X',
    'Y',
    'TYPE',
    'OBJECTID',
    'ID',
    'ZIP4',
    'TELEPHONE',
    'COUNTYFIPS',
    'NAICS_CODE',
    'NAICS_DESC',
    'SOURCE',
    'SOURCEDATE',
    'VAL_METHOD',
    'VAL_DATE',
    'WEBSITE',
    'STATE_ID',
    'ST_FIPS',
    'OWNER',
    'TTL_STAFF',
    'BEDS',
    'TRAUMA',
    'HELIPAD'
    ]

df = df.drop(to_drop, axis=1)

df
```

#### Remove closed hospitals

```python
df = df[df['STATUS'] == 'OPEN']

df = df[df['COUNTRY'] == 'USA']

states_to_keep = [ 'AL','AZ','AR','CA','CO','CT','DE','FL','GA','ID','IL','IN','IA','KS','KY','LA','ME','MD','MA','MI','MN','MS','MO','MT','NE','NV','NH',
                   'NJ','NM','NY','NC','ND','OH','OK','OR','PA','RI','SC','SD','TN','TX','UT','VT','VA','WA','WV','WI','WY' ]

df = df[df['STATE'].isin(states_to_keep)]
```

```python


df
```

```python
df = df.rename(columns={
    'NAME': 'Name',
    'CITY': 'City',
    'STATE': 'State',
    'LONGITUDE': 'Longitude',
    'LATITUDE': 'Latitude',
})

df = df[['Name', 'City', 'State', 'Longitude', 'Latitude']]

df
```

```python
df.to_csv('./data/Hospitals.csv', index=False)
```
