```python
import pandas as pd

pd.set_option('display.max_columns', None)

pd.set_option('display.max_rows', 100)

df = pd.read_csv('./raw_data/us-airports.csv')

df = df.drop(index=0)

df
```

#### Filter out minor airports

```python
df = df[df['type'].isin(['large_airport', 'medium_airport'])]

states_to_keep = [ 'AL','AZ','AR','CA','CO','CT','DE','FL','GA','ID','IL','IN','IA','KS','KY','LA','ME','MD','MA','MI','MN','MS','MO','MT','NE','NV','NH',
                   'NJ','NM','NY','NC','ND','OH','OK','OR','PA','RI','SC','SD','TN','TX','UT','VT','VA','WA','WV','WI','WY' ]

df = df[df['local_region'].isin(states_to_keep)]

df = df[~df['ident'].str.startswith('US-')]

df = df[~df['municipality'].str.contains('Fort')]

df = df[~df['name'].str.contains('Army|Base')]

df
```

```python
df['municipality'] = df['municipality'].str.upper()

df = df.rename(columns={
    'name': 'Name',
    'municipality': 'City',
    'local_region': 'State',
    'longitude_deg': 'Longitude',
    'latitude_deg': 'Latitude',
})

df = df[['Name', 'City', 'State', 'Longitude', 'Latitude']]

df
```

#### Remove unnecessary columns

```python
df.to_csv('./data/Alt_Airports.csv', index=False)
```
