```python
import pandas as pd

pd.set_option('display.max_columns', None)

pd.set_option('display.max_rows', 100)

df1 = pd.read_csv('./raw_data/GlobalAirportDatabase.csv')

df1
```

```python
df2 = pd.read_csv('./raw_data/BTS_Airports.csv')

df2
```

#### Filtering datasets

```python
df1 = df1[df1['Country'] == 'USA']

df1 = df1.dropna(subset=['Airport Name'])

to_keep = [
    'Airport Name',
    'City/Town',
    'Latitude Decimal Degrees',
    'Longitude Decimal Degrees'
]

df1 = df1[to_keep]

df1 = df1.rename(columns={'Airport Name': 'Name',
                          'City/Town': 'City',
                          'Latitude Decimal Degrees': 'Latitude',
                          'Longitude Decimal Degrees': 'Longitude',
                          })

df1
```

```python
df2 = df2[df2['Fac_Type'] == 'AIRPORT']

df2 = df2[~df2['State_Post'].isin(['AK', 'HI'])]

# These are mostly islands (not accessible for drivers)
df2 = df2.dropna(subset=['State_Post'])

# Drop unnecessary columns
to_keep = [
    'the_geom',
    'Fac_Name',
    'State_Post',
    'City'
]

df2 = df2[to_keep]

df2 = df2.rename(columns={'Fac_Name': 'Name',
                          'State_Post': 'State',
                          })

df2[['Longitude', 'Latitude']] = df2['the_geom'].str.extract(r'POINT \(([-\d\.]+) ([-\d\.]+)\)')

df2[['Longitude', 'Latitude']] = df2[['Longitude', 'Latitude']].astype(float)

df2 = df2.drop(labels='the_geom', axis=1)

df2
```

### Finding the intersection on Longitude and Latitude

This is mainly to filter out minor airports in `df1` while retaining the state and city names

```python
from geopy.distance import great_circle

def find_nearby_airports(df1, df2, radius=2):
    matches = []
    
    for _, row1 in df1.iterrows():
        lat1, lon1 = row1['Latitude'], row1['Longitude']
        
        for _, row2 in df2.iterrows():
            lat2, lon2 = row2['Latitude'], row2['Longitude']
            distance = great_circle((lat1, lon1), (lat2, lon2)).miles  # Calculate distance in miles
            
            if distance <= radius:
                matches.append({
                    'df1_Name': row1['Name'],
                    'df1_City': row1['City'],
                    'df2_Name': row2['Name'],
                    'df2_City': row2['City'],
                    'df2_State': row2['State'],
                    'Longitude': row2['Longitude'],
                    'Latitude': row2['Latitude'],
                })
    
    return pd.DataFrame(matches)

# Find matches within 1 mile
df = find_nearby_airports(df1, df2)

df
```

```python
# Formatting

df = df.rename(columns={
    'df1_Name': 'Name',
    'df2_City': 'City',
    'df2_State': 'State',
})

df = df[['Name', 'City', 'State', 'Longitude', 'Latitude']]

df
```

```python
df.to_csv('data/Airports.csv', index=False)
```
