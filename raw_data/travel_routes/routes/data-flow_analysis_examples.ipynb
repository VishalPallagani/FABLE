This is the code used to find examples for the data-flow analyses

```python
import pandas as pd

df = pd.read_csv('./output_data/summary_large.csv')

df
```

```python
def check_conditions(text):
    if isinstance(text, str):
        return 'Keep' in text and ('\'Turn right\'' in text or '\'Turn left\'' in text)
    return False

# List of columns to apply the check to
cols_to_check = ['plan_1', 'plan_2', 'plan_3']

# Create a boolean mask across the rows
mask = df[cols_to_check].applymap(check_conditions).any(axis=1)

# Filter the DataFrame
df = df[mask]

df = df.sort_values(by='average_plan_length')

df
```

```python

df.to_csv('./output_data/short_plans.csv')
```
